```python
import nibabel as nib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
dir = r"C:\Users\sarav\Documents\ADNI\ADNI Dataset"

def list_files(dir):
    r = []
    for root, dirs, files in os.walk(dir):
        for name in files:
            filepath = root + os.sep + name
            if filepath.endswith(".nii"):
                r.append(os.path.join(root, name))
    return r
```

```python
list_files_adni = list_files(dir)
```

```python
import plotly.express as px
```

```python
for num,x in enumerate(list_files_adni):
    img = nib.load(x)
    img_data = img.get_fdata()[:,:,100]
    plt.subplot(7,10,num+1)
    plt.axis('off')
    plt.imshow(img_data,cmap='gray')
```
